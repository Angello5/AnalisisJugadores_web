```python
import pandas as pd

file_path = "../data/fifa_players.csv"
fifa_data = pd.read_csv(file_path)
```

```python
fifa_data.isnull().sum()
```

```python
categorical_columnas = ['national_team','national_rating','national_team_position','national_jersey_number']
for col in categorical_columnas:
    fifa_data[col].fillna("No tiene equipo", inplace=True)
```

```python
fifa_data['value_euro'].fillna(fifa_data['value_euro'].median(), inplace=True)
fifa_data['wage_euro'].fillna(fifa_data['wage_euro'].median(), inplace=True)
fifa_data['release_clause_euro'].fillna(fifa_data['release_clause_euro'].median(), inplace=True)
```

```python
print(fifa_data.isnull().sum())
```

```python
#Eliminar columnas irrelevantes
fifa_data.drop(columns=['national_team','national_rating','national_team_position','national_jersey_number'], inplace=True)
```
